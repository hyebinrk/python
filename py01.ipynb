### format 함수를 사용한 포매팅

```python
# 바로 대입하기
print("I eat {0} of {1} apples".format(3,10))
```

```python
# 변수로 대입하기
total = 10
ea = 3
print("I eat {0} of {1} apples".format(ea,total))
```

```python
print("I eat {ea} of {total} apples".format(ea="three",total=10))
```

```python
# 인덱스와 이름을 혼용해서 넣기
print("I ate {0} apples. so I want sick for {day} days.".format(10, day=3))
```

```python
# 우측 정렬
print("{0:>10}".format("hi")) #띄어쓰기 포함해서 10자리 확보
```

```python
# 가운데 정렬
print("{0:^10}".format("hi"))
```

```python
# 좌측 정룔
print("{0:<10}".format("hi"))
```

```python
# 가운데 정렬 + 글자 채우기
print("{0:=^10}".format("hi"))
```

```python
# 가운데 정렬
print("{0:♡^10}".format("시온"))
```

```python
#소수점 표현하기
pi = 3.141592
print("{0:20.4f}".format(pi)) # 중괄호에 0만 넣었으면 문자가 되는거고 이런식으로 표현해야 숫자화됨
```

```python
#이스케이프 처리
print("{0}는 {{과}} 입니다.".format("중괄호"))
```

```python
# 2진수 처리
print("{0:b}".format(123))
print("{0:0>8b}".format(123))
```
