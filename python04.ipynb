```python
# max() //
def bs(li=[33,11,22,44]):
    sw = 0
    for x in range(len(li)):
        if sw < li[x]:
            sw = li[x]
    return sw
print(bs())
```

```python
# 테스트 : 함수 random 주사위
import random as r #random을 실행할 때마다 계속 random.random을 써야하므로 설정
import matplotlib.pyplot as plt
def dice(target = 250000):
    box = [] # 초기화 ⭐⭐⭐⭐⭐
    num = 0
    results = {}
    while num < target:
        num = num + 1
        ran = r.randint(1,6) #지정한 값까지 나옴 (1부터 6까지 나옴)
        box.append(ran)
    if target == len(box):
        print(f"{target}회 정상 진행되었습니다.")
    else:
        print(f"검증 결과 횟수 이상이 있습니다. 이유:({target}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
    for n in range(1,7):
        results[n] = box.count(n)
        print(f"{n}번이 {box.count(n)}번({(box.count(n)/len(box))*100:.2f}% 등장하였습니다.")
    # print(results)
    
    plt.bar(results.keys(),results.values(),color="pink")
    plt.show() 
```

```python
dice()
```

```python
#여러개의 입력값을 받는 함수 만들기
def add_many(*args):
    result = 0
    for i in args:
        result = result + i
    return result        
```

```python
res = add_many(1,4,5,2,4)
print(res)
```

```python
def add_mul(choice,*args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "sub":
        result =0
        for i in args:
            result = result - 1
    elif choice == "mul":
        result =1
        for i in args:
            result = result * 1
    elif choice == "div":
        result =1
        for i in args:
            result = result / 1
    return result
```

```python
add_mul('div',33,22,11)
```
