```python
import matplotlib.font_manager as fm
from matplotlib import rc
import matplotlib.pyplot as plt
path = './ref/D2Coding-Ver1.3.2-20180524.ttf'
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.text(0.2,0.4,'한글')
```

```python
import pandas as pd
file_path = "./ref/서울시안심이CCTV연계현황.csv"
ori = pd.read_csv(file_path,encoding='CP949')
ori.sample(5)
```

```python
ori = ori.rename(columns={'자치구':'location'})['location']
ori = ori.value_counts().reset_index()
ori = ori.set_index('location')
ori
```

```python
ori_barh = ori['count'].sort_values().plot(x = 'local', y = 'count', kind='barh', figsize=(13,9), ylabel='구', xlabel='CCTV 수', title = '서울 안심이 CCTV 설치현황')

plt.show()
```

- 강남구가 압도적으로 많고 그 다음으로 관악구, 서초구가 많고 설치 수의 편차가 구마다 큰 것으로 보인다.
- 가장 많은 강남구와 용산구의 CCTV 설치 수는 약 7000대의 차이가 난다.
