```python
from konlpy.tag import Okt

#한글 형태소 분석기
okt = Okt()

positive_words = ["좋다","행복","멋지다", "최고","사랑"]
negative_words = ["싫다","화나다","짜증","별로","나쁘다"]
text = input("당신의 기분을 한 문장으로 표현해주세요:")

words = okt.morphs(text, stem = True)

print(words)

p_mood = 0
n_mood = 0
for i in words:
    if i in positive_words:
        p_mood += 1
    elif i in negative_words:
        n_mood += 1
if p_mood > n_mood:
    print('당신의 기분은 좋습니다')
elif p_mood < n_mood:
    print('당신의 기분은 나쁩니다')
else:
    print('당신의 기분은 보통입니다.')

print(p_mood)
print(n_mood)
```

```python
# operator.attrgetter()
from operator import attrgetter

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    def pr(self):
        return self.name, self.age, self.grade

students = [
    Student('jane', 22, 'A'),
    Student('dave', 32, 'B'),
    Student('sally', 17, 'B'),
]

result = sorted(students, key=attrgetter('age'))
for r in result:
    print(r.pr())
```

## Threading

```python
import time

def long_tesk():
    for i in range(5):
        time.sleep(1)
        print(f"일하는 중...{i}")
long_tesk()
```

```python
# 기본작업 5초 X 5회 = 25초
start = time.time()
print("=====Start=====")
for n in range(5):
    long_tesk()
print("=====End=====")
print(time.time() - start)
```

```python
# thread를 사용
import threading
start = time.time()
print("=====Start=====")
threads = []
for n in range(5):
    t = threading.Thread(target=long_tesk)
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join()        
print("=====End=====")
print(time.time() - start)
```
