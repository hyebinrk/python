```python
import matplotlib.pyplot as plt  
from wordcloud import WordCloud  
import matplotlib.font_manager as fm  # matplotlib font_manager  import  .
from konlpy.tag import Komoran  # KoNLPy Komoran  import    .

font_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"  #    .

text = """      

    , ,  ,              .                  .    4  ,     (KHP)      18  .    11    ,      .        (BASA, Bilateral Aviation Safety Agreement) ,    (KC-100)  .
       

   ,          .     ,  , , , , , ,       ,       .     (Teal Group)     2023 125  ,       8 8,000   35%     .    IT           7       .  2023  5, 2027  3     .             (IT)       (PAV)   .         LTA(Lighter Than Air)  ,    .           .         ,  ,                 .          (EAV, Electrical Aerial Vehicle)              .     ,            .          (OPPAV, Optionally Piloted Personal Air Vehicle)   ,        (UTM, Unmanned Aerial System Traffic Management)            ,     ,        ."  #    ."""


#     .     .
# data = [t for t in text if bool(t.strip())]
# print(data)

# Komoran  .
komoran = Komoran()

#     .
nouns = komoran.nouns(text)

nouns_text = " ".join(nouns)

# WordCloud  .
# font_path    , width height      ,
#   . generate      .
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(text)

#  .
plt.figure(figsize=(10, 5))  #    10,  5 .
plt.imshow(wordcloud, interpolation="bilinear")  #     . 'bilinear' (interpolation)  .
plt.axis('off')  #  (x, y)   .
plt.savefig("word2.png")  #   'word2.png'  .
plt.show()  #   .
```
