```python
class Duck:
    def __init__(self,won,ea):
        self.won = won
        self.ea = ea
    def dduck(self):
        result = self.won * self.ea
        print(f'배가 고파서 떡볶이 {result}원어치 먹었다')
    def odeng(self):
        result = self.won * 0.3 * self.ea
        print(f'배가 고파서 오뎅 {result}원어치 먹었다')
    def sundae(self):
        result = self.won * 1.5 * self.ea
        print(f'배가 고파서 순대 {result}원어치 먹었다')
    
```

```python
# 테스트 7
class Pet:
    def __init__(self,name,species):
        """펫을 생성"""
        self.name = name
        self.species = species
        self.hunger = 50 #초기 배고픔 수치(0~100)
    def feed(self):
        """밥을 주면 배고픔 감소"""
        if self.hunger > 10:
            self.hunger -= 10
            return f"{self.name}에게 밥을 줬습니다.! 배고픔:{self.hunger}"
    def play(self):
        """놀아주면 배고픔이 증가"""
        if self.hunger <= 80:
            self.hunger += 20
        return f"{self.name}와 아주 잘 놀아줬어요!!!~~ 배고픔:{self.hunger}"
```

```python
# 펫 생성
dog1 = Pet("뽀삐","말티즈")
dog2 = Pet("아지","요크")
cat = Pet("컁이","나만고양이없어")
print(dog1.feed())
print(dog1.feed())
print(dog2.play())
print(dog2.play())
print(dog2.play())
print(dog1.play())
print(dog1.play())
print(cat.play())
print(cat.play())
print(cat.feed())
```

```python
# 테스트8
class Bacteria:
    def __init__(self,initial_population, growth_rate):
        """세균초기화"""
        self.population = initial_population
        self.growth_rate = growth_rate
        self.time = 0 # 시작시간
    def grow(self,hours):
        """일정 시간이 지나면 세균이 증식함"""
        self.time = hours
        self.population *= self.growth_rate**hours
        return f"{self.time}시간 후 세균 수 : {int(self.population)} 마리"
    def check(self):
        """현재 세균 수 확인"""
        return f"현재 세균 수 : {int(self.population)}마리"
```

```python
bacteria = Bacteria(100,2)
```

```python
print(bacteria.grow(1))
print(bacteria.grow(2))
print(bacteria.grow(3))
```

```python
# 테스트 9 은행예금 캡슐화 된 예제, 클래스 변수
class BankAccount:
    interest_rate = 0.02 #클래스변수
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance or 0    
    def bprint(self):
        return f"{self.owner}님의 잔고는{self.__balance}원 입니다."
    def deposit(self,amount):
        self.balance += amount
        return f"{self.owner}님의 계좌에 {amount}원이 입금 되었습니다. 현재 잔액:{self.__balance}원"
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance -= amount
            return f"{self.owner}님의 계좌에 {amount}원이 출금 되었습니다. 현재 잔액:{self.__balance}원"
        return "잔액이 부족합니다."
    def interest(self):
        total = self.__balance * (1 + self.interest_rate)
        return f"{self.interest_rate}%의 이자가 반영된 금액은 {total:.0f}원 입니다."

#인스턴스화
acc3 = BankAccount("철수",50000)
acc4 = BankAccount("영희",200000)

print(acc3.bprint())
print(acc4.bprint())
print(acc3.interest())
print(acc4.interest())
#클래스 변수 바꾸기
BankAccount.interest_rate = 0.03
print(acc3.interest())
print(acc4.interest())

#공격 후
#acc1.balance = 9000000
#print(acc3.balance)

#내장메소드로 캡슐화된 변수 접근
# print(acc3.deposit(30000))
# print(acc3.withdraw(5500))
```

```python
네임 맹글링 (Name Mangling)은 주로 객체 지향 프로그래밍에서 클래스의 멤버 변수를 프라이빗(private)으로 만들기 위해 사용되는 기술입니다. 파이썬에서는 두 개의 언더스코어(__)로 시작하는 변수 이름을 네임 맹글링을 통해 내부적으로 변환하여 클래스 외부에서 접근하지 못하도록 합니다.

다음은 네임 맹글링의 예시입니다:

class MyClass:
    def init(self, value):
        self.__private_var = value  # 프라이빗 변수

    def get_private_var(self):
        return self.__private_var

# 인스턴스 생성
obj = MyClass(42)

# 프라이빗 변수 접근 (이 방법은 실패합니다)
# print(obj.__private_var)  # AttributeError 발생

# 네임 맹글링을 통해 접근 (이 방법은 성공합니다)
print(obj._MyClass__private_var)  # 출력: 42

# 메서드를 통해 접근
print(obj.get_private_var())  # 출력: 42
```

```python
class VendingMachine:
    def __init__(self):
        """자동판매기 초기화"""
        self.balance = 0 #사용자의 잔액
        self.item = { #상품 목록 (이름: 가격)
            "콜라": 1500,
            "사이다": 1400,
            "물": 1000
        }
    def insert_money(self, amount):
        """잔액 충전"""
        self.balance += amount
        return f"잔액이 {self.balance}원으로 증가했습니다."
    def buy_item(self, item_name):
        """상품구매"""
        if item_name not in self.item:
            return f"{item_name}은(는) 판매하지 않는 상품입니다."
        price = self.item[item_name]
        if self.balance < price:
            return f"잔액이 부족합니다! ({item_name} 가격: {price}원, 현재 잔액: {self.balance}원"
        self.balance -= price
        return f"{item_name}을(를) 구매했습니다! 남은 잔액: {self.balance}원"
    def refund(self):
        """잔액 반환"""
        amount = self.balance
        self.balance = 0
        return f"잔액 {amount}원이 반환되었습니다."
#자동판매기 객체 생성
vm = VendingMachine()
# 잔액 충전
print(vm.insert_money(2000))
# 상품 구매
print(vm.buy_item("콜라")) # 콜라을(를) 구매했습니다! 남은 잔액: 500원"
print(vm.buy_item("사이다")) # 잔액이 부족합니다! (사이다 가격: 1400원, 현재 잔액: 500원)
#잔액 반환
print(vm.refund()) # 잔액 500원이 반환되었습니다.
```
