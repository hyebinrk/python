```python
'''
ë„¤ì„ ë§¹ê¸€ë§ (Name Mangling)ì€ ì£¼ë¡œ ê°ì²´ ì§€í–¥ í”„ë¡œê·¸ë˜ë°ì—ì„œ í´ë˜ìŠ¤ì˜ ë©¤ë²„ ë³€ìˆ˜ë¥¼ í”„ë¼ì´ë¹—(private)ìœ¼ë¡œ ë§Œë“¤ê¸° ìœ„í•´ ì‚¬ìš©ë˜ëŠ” ê¸°ìˆ ì…ë‹ˆë‹¤. íŒŒì´ì¬ì—ì„œëŠ” ë‘ ê°œì˜ ì–¸ë”ìŠ¤ì½”ì–´(__)ë¡œ ì‹œì‘í•˜ëŠ” ë³€ìˆ˜ ì´ë¦„ì„ ë„¤ì„ ë§¹ê¸€ë§ì„ í†µí•´ ë‚´ë¶€ì ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ í´ë˜ìŠ¤ ì™¸ë¶€ì—ì„œ ì ‘ê·¼í•˜ì§€ ëª»í•˜ë„ë¡ í•©ë‹ˆë‹¤.

ë‹¤ìŒì€ ë„¤ì„ ë§¹ê¸€ë§ì˜ ì˜ˆì‹œì…ë‹ˆë‹¤:
'''

class MyClass:
    def init(self, value):
        self.__private_var = value  # í”„ë¼ì´ë¹— ë³€ìˆ˜

    def get_private_var(self):
        return self.__private_var

# ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
obj = MyClass(42)

# í”„ë¼ì´ë¹— ë³€ìˆ˜ ì ‘ê·¼ (ì´ ë°©ë²•ì€ ì‹¤íŒ¨í•©ë‹ˆë‹¤)
# print(obj.__private_var)  # AttributeError ë°œìƒ

# ë„¤ì„ ë§¹ê¸€ë§ì„ í†µí•´ ì ‘ê·¼ (ì´ ë°©ë²•ì€ ì„±ê³µí•©ë‹ˆë‹¤)
print(obj._MyClass__private_var)  # ì¶œë ¥: 42

# ë©”ì„œë“œë¥¼ í†µí•´ ì ‘ê·¼
print(obj.get_private_var())  # ì¶œë ¥: 42
```

```python
class VendingMachine:
    def __init__(self):
        """ìë™íŒë§¤ê¸° ì´ˆê¸°í™”"""
        self.balance = 0 #ì‚¬ìš©ìì˜ ì”ì•¡
        self.item = { #ìƒí’ˆ ëª©ë¡ (ì´ë¦„: ê°€ê²©)
            "ì½œë¼": 1500,
            "ì‚¬ì´ë‹¤": 1400,
            "ë¬¼": 1000
        }
    def insert_money(self, amount):
        """ì”ì•¡ ì¶©ì „"""
        self.balance += amount
        return f"ì”ì•¡ì´ {self.balance}ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤."
    def buy_item(self, item_name):
        """ìƒí’ˆêµ¬ë§¤"""
        if item_name not in self.item:
            return f"{item_name}ì€(ëŠ”) íŒë§¤í•˜ì§€ ì•ŠëŠ” ìƒí’ˆì…ë‹ˆë‹¤."
        price = self.item[item_name]
        if self.balance < price:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤! ({item_name} ê°€ê²©: {price}ì›, í˜„ì¬ ì”ì•¡: {self.balance}ì›"
        self.balance -= price
        return f"{item_name}ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡: {self.balance}ì›"
    def refund(self):
        """ì”ì•¡ ë°˜í™˜"""
        amount = self.balance
        self.balance = 0
        return f"ì”ì•¡ {amount}ì›ì´ ë°˜í™˜ë˜ì—ˆìŠµë‹ˆë‹¤."
#ìë™íŒë§¤ê¸° ê°ì²´ ìƒì„±
vm = VendingMachine()
# ì”ì•¡ ì¶©ì „
print(vm.insert_money(2000))
# ìƒí’ˆ êµ¬ë§¤
print(vm.buy_item("ì½œë¼")) # ì½œë¼ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡: 500ì›"
print(vm.buy_item("ì‚¬ì´ë‹¤")) # ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤! (ì‚¬ì´ë‹¤ ê°€ê²©: 1400ì›, í˜„ì¬ ì”ì•¡: 500ì›)
#ì”ì•¡ ë°˜í™˜
print(vm.refund()) # ì”ì•¡ 500ì›ì´ ë°˜í™˜ë˜ì—ˆìŠµë‹ˆë‹¤.
```

```python
import random

class Character:
    def __init__(self, name, hp, attack_p):
        """ìºë¦­í„° ì´ˆê¸°í™”"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self, opponent):
        """ìƒëŒ€ë°©ì„ ê³µê²©"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}ê°€ {opponent.name}ë¥¼ ê³µê²©í•˜ì—¬ {damage}ë°ë¯¸ì§€ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤.")
    def now_status(self):
        """í˜„ì¬ ì²´ë ¥ ìƒíƒœ í™•ì¸"""
        return f"{self.name} ì²´ë ¥:{self.hp}"
    def is_alive(self):
        """ìºë¦­í„° ìƒì¡´ì—¬ë¶€ í™•ì¸"""
        return self.hp > 0
```
