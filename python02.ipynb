### 리스트의 인덱싱과 슬라이싱

```python
odd = [1,3,5,7,9]
a = []
b = [1,2,3]
c = ['Life','is','too','short']
d= [1,2,'Life','is']
e = [1,2,['Life','is']]
print(e[2][0])
print(e[2][1])
```

```python
# 리스트의 인덱싱과 연산
print(b[1] + b[2])
print(c[0] + c[3])
print(c[0] * b[2])
```

```python
# 리스트의 슬라이싱
a = [11,22,33,44,55]
print(a[0:2])
print(a[2:])
print(a[:-3])
print(a[-2:])
print(a[::2])
print(a[::-2])
```

### 리스트 연산,수정,삭제하기

```python
a = [11,22,33]
b = [44,55,66]
#리스트 더하기
print(a+b)
#리스트 반복하기
print(a * 3)
#리스트 길이 구하기
c = a*3
len(c)
```

```python
#리스트 값 수정하기
a = [11,22,33]
a[2] = 44
print(a)
# a[3] = 55 # 없는 요소를 수정하는 것은 에러 print(a)해도 안 나옴
# a[1] = '' #빈칸 수정 삽입
# a[1] = 0 # 0으로 수정 삽입
#리스트 값 삭제하기
del a [1]
print(a)
```

```python
#리스트 슬라이싱으로 삭제
print(c)
del c[:4]
print(c)
```

```python
# 리스트정렬 원본 유지하기
aa = [66,77,88,99,0]
bb = sorted(aa)
cc = sorted(aa,reverse=True)
print(aa)
print(bb)
print(cc)
```

```python
# sorted 응용
fruit = ['banana','kiwi','orange','melon','pear','apple','strawberry','blueberry']
result = sorted(fruit,key=len)
result2 = sorted(fruit,key=len,reverse=True)
print(fruit)
print(result)
print(result2)
```

```python
print('과일을 먹어볼까?')
for r in result:
    if r == 'banana':
        print(r+'못먹겠어 웩!')
    if r != 'banana':
     print(r + ' 냠냠 맛있다')
print('와 ~ 다 먹었다')
```

```python
#인덱스 반환 - index
fruit = ['banana','kiwi','orange','melon','pear','apple','strawberry','blueberry']
fruit.index('melon')
# fruit.index('m') 와 같이 없는 문자를 넣으면 오류가 뜬다
# fruit.find('m') 을 치면 없는 문자라 -1이 뜰 것 같지만 이런건 없다.-오류
```

```python
#인덱스 응용
fruit = ['banana','kiwi','orange','melon','pear','apple','strawberry','blueberry']
want = input('먹고 싶은 과일을 입력하세요.')
where = fruit.index(want)
print(f'{want}를 좋아하시는군요!!! 그것은 {where + 1}번째에 있습니다.')
```
