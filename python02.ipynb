### 리스트 연산,수정,삭제하기

```python
a = [11,22,33]
b = [44,55,66]
#리스트 더하기
print(a+b)
#리스트 반복하기
print(a * 3)
#리스트 길이 구하기
c = a*3
len(c)
```
