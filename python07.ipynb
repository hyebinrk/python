```python
from konlpy.tag import Okt

#한글 형태소 분석기
okt = Okt()

positive_words = ["좋다","행복","멋지다", "최고","사랑"]
negative_words = ["싫다","화나다","짜증","별로","나쁘다"]
text = input("당신의 기분을 한 문장으로 표현해주세요:")

words = okt.morphs(text, stem = True)

print(words)

p_mood = 0
n_mood = 0
for i in words:
    if i in positive_words:
        p_mood += 1
    elif i in negative_words:
        n_mood += 1
if p_mood > n_mood:
    print('당신의 기분은 좋습니다')
elif p_mood < n_mood:
    print('당신의 기분은 나쁩니다')
else:
    print('당신의 기분은 보통입니다.')

print(p_mood)
print(n_mood)
```
