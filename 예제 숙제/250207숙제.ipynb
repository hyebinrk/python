```python
# 참고용 1
ls = []
la = []
how = 100
for h in range(how):
    lotto = sorted(random.sample(range(1,46),6))
    la.append(lotto)
    ls = ls + lotto
# print(la, type(la))
ss = []
for n in range(1,46):
    ss.append(ls.count(n))
    print(f'{n}이 출현한 횟수는 {ls.count(n)}회 입니다.')
print(ss,sum(ss) == how*6)
if sum(ss) == (how * 6):
    print(f'출현횟수에 {how}대한 검증결과 이상없습니다.')
```
