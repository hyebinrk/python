```python
import pandas as pd
file_path = "./ref/청소년정신건강20250304.csv"
ori = pd.read_csv(file_path)
ori
```

```python
ori.columns = ori.iloc[0]
ori.iloc[1].reset_index()
ori
```

```python
#컬럼명 지정
new_name = ['시점','스트레스','스트레스남학생','스트레스여학생','우울감경험률','우울남학생','우울여학생','자살생각률','자살남학생','자살여학생']
ori.columns = new_name
ori
```

```python
# 시점을 없애고, 0번 행도 없앰
ori = ori.loc[1:,'스트레스':]
ori
```

```python
# 2번 인덱스 행을 생성, 100에서 값을 뺀 값
ori.loc[2] = 100 - ori.loc[1,'스트레스':].astype(float)
# 마지막에 응답컬럼을 생성, 그렇다 or 아니다.
ori['응답'] = ['그렇다','아니다']
ori
```

```python
ori = ori.set_index('응답')
ori
```

```python
# OS에 사용가능한, 설치된 폰트를 나열
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
fontlist = fm.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
path = 'C:\\Windows\\Fonts\\Hancom Gothic Bold.ttf'
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.text(0.2,0.4,'한글')
```

```python
ori['스트레스'].astype(float).plot.pie(explode = [0,0.02])
```

```python
fig, ax = plt.subplots(1,3,figsize=(10,8))

ori['스트레스'].astype(float).plot.pie(explode = [0,0.1], ax = ax[0] , autopct="%.1f%%")
ax[0].set_title('스트레스를 받은 적이 있다.')
ax[0].set_ylabel('')

ori['우울감경험률'].astype(float).plot.pie(explode = [0,0.1], ax = ax[1] , autopct="%.1f%%")
ax[1].set_title('우울증을 경험한 적이 있다')
ax[1].set_ylabel('')

ori['자살생각률'].astype(float).plot.pie(explode = [0,0.1], ax = ax[2] , autopct="%.1f%%")
ax[2].set_title('자살을 고민한 적이 있다.')
ax[2].set_ylabel('')

plt.show()
```
